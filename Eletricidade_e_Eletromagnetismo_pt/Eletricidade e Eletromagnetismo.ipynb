```python
import numpy as np  # Para mais detalhes https://numpy.org/
import matplotlib.pyplot as plt # Para mais detalhes https://matplotlib.org/
```

```python
ri = 0.
rf = 5.
dr = 0.1
r = np.arange(ri,rf,dr)
print(r)
```

```python
r = np.arange(ri,rf+dr,dr)
print(r)
```

```python
def F_e(r,q1,q2):
    k = 9*10**9
    e2= (1.6*10**(-19.))**2
    return k*q1*q2*e2/(r**2)
```

```python
q1 = -1. # carga q_1
q2 = -1. # carga q_2

plt.plot(r,F_e(r,q1,q2),'-x')
plt.xlabel(r'r (m)')
plt.ylabel(r'$F_{e}(r)$ (N)')

plt.show()
```

```python
# Corrija o erro da divisao por zero:
ri = 0.001
dr = 0.001
rf = 5.
r = np.arange(ri,rf+dr,dr)
print(r)
```

```python
q1 = -1. # carga q_1
q2 = -1. # carga q_2

plt.plot(r,F_e(r,q1,q2),'-x')
plt.xlabel(r'r (m)')
plt.ylabel(r'$F_{e}(r)$ (N)')
plt.xlim(0,0.01)
plt.show()
```

```python
plt.plot(r,F_e(r,q1,q2)/(9*10**9),'-')
plt.xlabel(r'r (m)')
plt.ylabel(r'$F_{e}(r)$ (N)')
plt.yscale('log')
plt.show()
```
